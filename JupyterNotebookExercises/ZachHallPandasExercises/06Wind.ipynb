### Step 1. Import the necessary libraries

```python
import pandas as pd
import datetime
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data)

### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
data = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data', 
                   delim_whitespace=True)
def datefunc(row):
    row['Date'] = datetime.datetime(2000 + int(row.Yr), int(row.Mo), int(row.Dy))
    return row
datefunc(data.iloc[0])
data = data.apply(datefunc, axis=1)
data = data.drop(['Yr', 'Mo', 'Dy'], axis=1)
data
```

### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
def datefix(row):
    row['Date'] = datetime.datetime(row['Date'].year-100, row['Date'].month, row['Date'].day)
    return row

data = data.apply(datefix, axis=1)
data
```

### Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
data = data.set_index('Date')
data
```

### Step 6. Compute how many values are missing for each location over the entire record.  
#### They should be ignored in all calculations below. 

```python
missing = data.isnull().values.sum()
print(missing)
```

### Step 7. Compute how many non-missing values there are in total.

```python
nonmissing = (len(data)*len(data.columns))-missing
print(nonmissing)
```

### Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
#### A single number for the entire dataset.

```python
speeds = data.values
fixedspeeds = []
for row in speeds:
    fixedspeeds += list(row[~np.isnan(row)])
fixedspeeds = np.array(fixedspeeds)
print(fixedspeeds.mean())
```

### Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days 

#### A different set of numbers for each location.

```python
places = data.columns
mins = [data[col].dropna().min() for col in places]
maxes = [data[col].dropna().max() for col in places]
means = [data[col].dropna().mean() for col in places]
stds = [data[col].dropna().std() for col in places]
loc_stats = pd.DataFrame(np.array([mins, maxes, means, stds]).T, index = places, columns = ['min', 'max', 'mean', 'std'])
loc_stats
```

### Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.

#### A different set of numbers for each day.

```python
mins = [data.iloc[i].dropna().min() for i in range(len(data))]
maxes = [data.iloc[i].dropna().max() for i in range(len(data))]
means = [data.iloc[i].dropna().mean() for i in range(len(data))]
stds = [data.iloc[i].dropna().std() for i in range(len(data))]
day_stats = pd.DataFrame(np.array([mins, maxes, means, stds]).T, index=data.index, columns = ['min', 'max', 'mean', 'std'])
day_stats
```

### Step 11. Find the average windspeed in January for each location.  
#### Treat January 1961 and January 1962 both as January.

```python
janwindspeeds = []
places = data.columns
for place in places:
    janwindspeeds.append((place, data[data.index.month==1][place].dropna().mean()))
janwindspeeds
```

### Step 12. Downsample the record to a yearly frequency for each location.

```python
years = list(set(data.index.year))
means = [[data.loc[str(year)][place].dropna().mean() for year in years] for place in places]
year_stats = pd.DataFrame(np.array(means).T, index=years, columns = places)
year_stats
```

### Step 13. Downsample the record to a monthly frequency for each location.

```python
months = [i for i in range(1, 13)]
means = [[data[(data.index.year==year)&(data.index.month==month)][place].dropna().mean() for year in years for month in months]
        for place in places]
periods = ['{}/{}'.format(year, month) for year in years for month in months]
month_stats = pd.DataFrame(np.array(means).T, index=periods, columns=places)
month_stats
```

### Step 14. Downsample the record to a weekly frequency for each location.

```python
weeks = [i for i in range(1, 53)]
means = [[data[(data.index.year==year)&(data.index.isocalendar().week==week)][place].dropna().mean() for year in years for week in weeks]
        for place in places]
periods = ['{} week {}'.format(year, week) for year in years for week in weeks]
week_stats = pd.DataFrame(np.array(means).T, index=periods, columns=places)
week_stats
```

### Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
mins = []
maxes = []
means = []
stds = []
for week in weeks:
    weeklyspeeds = []
    speeds = data[(data.index.year==1961)&(data.index.isocalendar().week==week)].values
    for row in speeds:
        weeklyspeeds += list(row[~np.isnan(row)])
    weeklyspeeds = np.array(weeklyspeeds)
    mins.append(weeklyspeeds.min())
    maxes.append(weeklyspeeds.max())
    means.append(weeklyspeeds.mean())
    stds.append(weeklyspeeds.std())
speeds1961 = pd.DataFrame(np.array([mins, maxes, means, stds]).T, index = weeks, columns = ['min', 'max', 'mean', 'std'])
speeds1961
```
