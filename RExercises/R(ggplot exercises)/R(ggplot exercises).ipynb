```r
library(ggplot2)
library(readr)
dat <- read_csv('EconomistData.csv')
```

```r
ggplot(dat, aes(CPI, HDI)) + geom_point()
```

```r
ggplot(dat, aes(CPI, HDI)) + geom_point(colour='blue')
```

```r
ggplot(dat, aes(CPI, HDI, colour=Region)) + geom_point()
```

```r
ggplot(dat, aes(CPI, HDI, colour=Region)) + geom_point(size=2)
```

```r
ggplot(dat, aes(CPI, HDI, colour=Region)) + geom_point(aes(size=HDI))
```

```r
ggplot(dat, aes(CPI, HDI)) + geom_point() + geom_smooth()
```

```r
ggplot(dat, aes(CPI, HDI)) + geom_point() + geom_smooth(method=lm)
```

```r
ggplot(dat, aes(CPI, HDI)) + geom_point() + geom_line(stat='smooth')
```

```r
ggplot(dat, aes(CPI, HDI, colour=Region)) + geom_point() + xlab('Consumer price index') + ylab('Human development index')
```

```r
ggplot(dat, aes(CPI, HDI, colour=Region)) + geom_point() + xlab('Consumer price index') + ylab('Human development index') + scale_colour_brewer(palette='Accent')
```
